### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
df['item_price_num'] = df['item_price'].str.replace("$", "", regex=False).astype(float)
```

```python
num_items_order = df.groupby('order_id')[['quantity']].sum()

def fun_aux(x):
    x['total_amount'] = (x['quantity'] * x['item_price_num']).sum()
    return x
price_per_order = df.groupby('order_id')[['order_id','quantity', 'item_price_num']].apply(fun_aux)
```

```python
price_per_order[['order_id', 'total_amount']].drop_duplicates()
```

```python
df_plot = pd.merge(num_items_order, price_per_order[['order_id', 'total_amount']].drop_duplicates(), on='order_id')
```

```python
df_plot.dtypes
```

```python
df_plot.plot(x='total_amount', y='quantity', kind='scatter')
```

```python
df_sorted_by_amount = df_plot.sort_values(by='total_amount', ascending=False)
ls_top5_ids = df_sorted_by_amount.head()['order_id'].unique()
df_notbig = df_plot[~df_plot['order_id'].isin(ls_top5_ids)]
df_notbig.plot(x='total_amount', y='quantity', kind='scatter')
```
